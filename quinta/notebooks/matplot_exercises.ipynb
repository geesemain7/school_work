```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, num=100)
fx = 2*x + 1

fig, ax = plt.subplots()
ax.plot(x, fx, label="f(x) = 2x + 1")

ax.set_title("Esercizio 1: grafico lineare")
ax.set_xlabel("x")
ax.grid(True)
ax.legend()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, num=100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, label="y = sin(x)", linestyle='--', color='r')
ax.plot(x, y2, label="y = cos(x)", linestyle=':', color='g')
ax.set_title("Confronto tra sin(x) e cos(x)")
ax.set_xlabel("x (radianti)")
ax.legend()
plt.show()
```

```python
x = np.random.uniform(0, 10, size=100)
y = np.random.uniform(0, 10, size=100)

colors = np.random.uniform(0, 1, size=(100,4))
diameter = np.random.uniform(5, 50, size=100)

fig, ax = plt.subplots()
ax.scatter(x, y, c=colors, s=diameter, marker='o')

ax.set_title("Grafico a dispersione")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
d1 = np.random.uniform(0, 1, size=100)

fig,ax = plt.subplots()
ax.hist(d1, bins = 10, color = 'b')

ax.set_xlabel("Valore")
ax.set_ylabel("Freuqenza")
ax.set_xlim(xmin = 0, xmax = 1.0)
ax.set_title("Istogramma di una Distribuzione Uniforme")
plt.show()
```
