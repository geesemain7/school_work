```python
from sympy import *
from sympy.abc import *

plt_limit = [-6, 6]

init_printing()
```

# **Livello del Re**

## **Esercizio 1**

```python
f = 1/x
```

```python
limit(f, x, 0, dir='-')     # limite destro
```

```python
limit(f, x, 0, dir='+')     # limite sinistro
```

```python
plot(f, xlim=plt_limit, ylim=plt_limit)
```

### *Cosa significa "da destra" e "da sinistra?"*

Per "da sinistra", si intende la limite di $x$ che tende a $0^-$:
### $$\lim\limits_{x \to 0^-}\frac{1}{x}$$

Invece per "da destra", si intende la limite di $x$ che tende a $0^+$:
### $$\lim\limits_{x \to 0^+}\frac{1}{x}$$


### *I due limiti sono uguali? Che tipo di discontinuità ha questa funzione in 0?*

No, $\lim\limits_{x \to 0^-} = -\infty$ e $\lim\limits_{x \to 0^+} = +\infty$. Questo è perché si tratta di una discontinuità asintotica.


### *In quali situazioni reali potresti voler sapere "da che lato" ti avvicini a un punto? Perché può servire saperlo?*

Un esempio può essere nell'analisi della termodinamica, dove il comportamento di una funzione può dare informazioni importanti, ad esempio temperature critiche nel contesto delle tranzisioni di fase dei materiali.

## **Esercizio 2**

```python
fd = diff(x**2, x)
fd
```

```python
gd = diff(cos(x), x)
gd
```

```python
hd = diff(ln(x), x)
hd
```

```python
graph = plot(fd, gd, hd, xlim=plt_limit, ylim=plt_limit, legend=True, show=False)

graph[0].label = "f'"
graph[1].label = "g'"
graph[2].label = "h'"

graph.show()
```

### *Che cosa rappresenta la derivata in ciascun caso?*

Una funzione derivata $f'(x)$ che descrive la coefficiente angolare di tutti i punti *(escluso i punti di non-derivabilità)* sulla sua funzione $f(x)$.


### *Quale di queste funzioni cresce più rapidamente?*

La funzione che cresce più rapidamente sarà quello con il coefficiente angolare maggiore. In questo caso analizzando le funzioni derivate, possiamo determinare che $f(x)$ è la funzione che cresce più rapidamente delle altre.


### *Come cambia il comportamento della funzione in base al segno della derivata?*

Siccome il segno di una funzione derivata indica la monotonia di una funzione, indica in quali intervalli la funzione è discendente o ascendente.

# **Livello delle Arene**

## **Esercizio 2 - Studio del comportamento di $\frac{|x|}{x}$**

```python
f = Abs(x)/x
```

```python
limit(f, x, 0, dir='-')
```

```python
limit(f, x, 0, dir='+')
```

```python
plot(f, xlim=plt_limit, ylim=plt_limit)
```

### *I due limiti coincidono?*

No, $\lim\limits_{x \to 0^-}\frac{|x|}{x} = -1$ e $\lim\limits_{x \to 0^+}\frac{|x|}{x} = 1 $.


### *La funzione è continua in 0?*

No, $x = 0$ è un punto salto perché $\lim\limits_{x \to 0^-}\frac{|x|}{x} \neq \lim\limits_{x \to 0^+}\frac{|x|}{x}$: i risultati sono numeri reali ma non coincidenti.


### *Come rappresenteresti graficamente questa funzione vicino a 0?*

Si rappresenta come proprio un punto salto: non c'è nessuna connessione tra le coordinate $(0; -1)$ e $(0; 1)$.
