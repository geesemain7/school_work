```python
from sympy import *
from sympy.abc import *

plt_limit = [-7, 7]

init_printing()
```

# **Livello del Re**

## **Esercizio 1**

```python
f = 1/x
```

```python
limit(f, x, 0, dir='-')     # limite destro
```

```python
limit(f, x, 0, dir='+')     # limite sinistro
```

```python
plot(f, xlim=plt_limit, ylim=plt_limit, xlabel=None, ylabel=None,)
```

### *Cosa significa "da destra" e "da sinistra?"*

Per "da sinistra", si intende la limite di $x$ che tende a $0^-$:
### $$\lim\limits_{x \to 0^-}\frac{1}{x}$$

Invece per "da destra", si intende la limite di $x$ che tende a $0^+$:
### $$\lim\limits_{x \to 0^+}\frac{1}{x}$$


### *I due limiti sono uguali? Che tipo di discontinuità ha questa funzione in 0?*

No, $\lim\limits_{x \to 0^-} = -\infty$ e $\lim\limits_{x \to 0^+} = +\infty$. Questo è perché si tratta di una discontinuità asintotica.


### *In quali situazioni reali potresti voler sapere "da che lato" ti avvicini a un punto? Perché può servire saperlo?*

Un esempio può essere nell'analisi della termodinamica, dove il comportamento di una funzione può dare informazioni importanti, ad esempio temperature critiche nel contesto delle tranzisioni di fase dei materiali.

### *Che cosa rappresenta la derivata in ciascun caso?*

Una funzione derivata $f'(x)$ che descrive la coefficiente angolare di tutti i punti *(escluso i punti di non-derivabilità)* sulla sua funzione $f(x)$.


### *Quale di queste funzioni cresce più rapidamente?*

La funzione che cresce più rapidamente sarà quello con il coefficiente angolare maggiore. In questo caso analizzando le funzioni derivate, possiamo determinare che $f(x)$ è la funzione che cresce più rapidamente delle altre.


### *Come cambia il comportamento della funzione in base al segno della derivata?*

Siccome il segno di una funzione derivata indica la monotonia di una funzione, indica in quali intervalli la funzione è discendente o ascendente.

# **Livello delle Arene**

## **Esercizio 2 - Studio del comportamento di $\frac{|x|}{x}$**

```python
f = Abs(x)/x
```

```python
limit(f, x, 0, dir='-')
```

```python
limit(f, x, 0, dir='+')
```

```python
plot(f, xlim=plt_limit, ylim=plt_limit)
```

### *I due limiti coincidono?*

No, $\lim\limits_{x \to 0^-}\frac{|x|}{x} = -1$ e $\lim\limits_{x \to 0^+}\frac{|x|}{x} = 1 $.


### *La funzione è continua in 0?*

No, $x = 0$ è un punto salto perché $\lim\limits_{x \to 0^-}\frac{|x|}{x} \neq \lim\limits_{x \to 0^+}\frac{|x|}{x}$: i risultati sono numeri reali ma non coincidenti.


### *Come rappresenteresti graficamente questa funzione vicino a 0?*

Si rappresenta come proprio un punto salto: non c'è nessuna connessione tra le coordinate $(0; -1)$ e $(0; 1)$.

## **Esercizio 2 - Derivate composte**

```python
fd = diff(exp(x) * sin(x), x)
fd
```

```python
gd = diff(x * ln(x), x)
fd
```

```python
graph = plot(fd, gd, xlim=plt_limit, ylim=plt_limit, xlabel=None, ylabel=None, legend=True, show=False)

graph[0].label = "f'"
graph[1].label = "g'"

graph.show()
```

### *In che modo il prodotto tra due funzioni influenza la derivata?*

La derivata viene influenzata dal prodotto tra due funzioni, perché data due funzioni $f(x)$ e $g(x)$, la formula della derivata diventa:
#### $D[f(x) · g(x)] = f'(x) · g(x) + f(x) · g'(x)$


### *Cosa succede al comportamento della funzione nei punti dove una delle due varia velocemente?*

Mentre il coefficiente angolare di $f(x)$ aumenta in modo sempre più ampia, mentre quella di $g(x)$ diminuisce.

# **Livello Elite**

## **Esercizio 1 - Punto di massimo**

```python
f = -x**2 + 4*x
fd = diff(f, x)    # Calcolo della derivata
solve(fd, x)    # Calcolo del punto stazionario
```

### *Che tipo di punto è? Minimo o massimo?*

Il punto di $f(x)$ a $x = 2$ è un massimo, perché $f(x)$ è una parabola rivolta verso il basso grazie al fatto che il suo termine di secondo grado è negativo.


### *Come fai a capirlo senza il grafico?*

Perché $f(x)$ è una parabola di tipo $y = ax^2 + bx + c$ rivolta verso il basso, grazie al fatto che il suo termine di secondo grado è negativo $(a = -1)$. Le parabole sono dotate di o un massimo o un minimo, *mai* entrambi. Quindi si può dire con certezza senza alcun calcolo che il punto stazionario su $(x = 2)$ è un massimo. 

Infatti se guardiamo il grafico...

```python
graph = plot(f, fd, xlim=plt_limit, ylim=plt_limit, xlabel=None, ylabel=None, legend=True, show=False)

graph[0].label = "f"
graph[1].label = "f'"

graph.show()
```

...questa tesi è verificata.

### *In quale contesto reale potrebbe servirti trovare un massimo?*

I massimi e minimi sono fondamentali in certi contesti reali, ad esempio: nella ricerca di un rapporto tra spinta e consumo energetico di una turboventola per gli aeri, con lo scopo di trovare le condizioni in qui c'è una resa massima di spinta con il minor consumo energetico possibile.

## **Esercizio 2 - Grafico di funzione e derivata**

```python
f = x * exp(-x)
fd = diff(f, x)

graph = plot(f, fd, xlim=plt_limit, ylim=plt_limit, xlabel=None, ylabel=None, legend=True, show=False)

graph[0].label = "f"
graph[1].label = "f'"

graph.show()
```

### *Dov'è positiva la derivata? E dov'è negativa?*

La derivata $f'(x)$ è positiva nell'intervallo $]-\infty; 1]$ e negativa nell'intervallo $[1; \infty[$.


### *In quali punti la derivata si annulla? Che cosa succede alla funzione in quei punti?*

C'è solo un punto dove $f'(x) = 0$, e in quel punto lì cambia al monotonia di $f(x)$ da ascendente a discendente.

$f'(x)$ *sembra* di avere valore $0$ dopo un certo punto, ma non tocca mai l'ascissa. Questo è perché se cerchiamo di trovare un asintoto orrizontale calcolando...
## $$\lim\limits_{x \to +\infty}f'(x)$$

```python
limit(fd, x, +oo)
```

... scopriamo che esiste un asintoto con equazione $(y = 0)$.

## *Che cosa ti racconta la derivata sul comportamento della funzione?*

La derivata ci puo indicare la monotonia attraverso il suo segno *(quindi anche informazioni sui punti stazionari)*. Però possiamo ricavare un'altro dato usando la derivata seconda $f''(x)$:

```python
fd2 = diff(fd, x)
solve(fd2, x)
```

Il risultato di solve() ci restituisce 2, qundi $f''(x)$ interseca l'ascissa in $x = 2$. Questo ci indica che $f(x)$ ha un punto di flesso in $x = 2$. Infatti se lo andiamo a vedere...

```python
graph = plot(f, fd2, xlim=[0, 3], ylim=[-1.5, 1.5], xlabel=None, ylabel=None, legend=True, show=False)

graph[0].label = "f(x)"
graph[1].label = "f''(x)"

graph.show()
```

... $f(x)$ cambia concavità da verso il basso a verso l'alto dopo $x = 2$.
