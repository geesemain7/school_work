```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 200)
F1 = x
F2 = x**2
F3 = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, F1, label="Funzione F1")
ax.plot(x, F2, label="Funzione F2")
ax.plot(x, F3, label="Funzione F3")

ax.set_title("Il mio primo grafico")
ax.set_xlabel("Ascissa")
ax.set_ylabel("Ordinata")
ax.legend()
plt.show()
```

```python
y = np.array([1, -2, 0, 3, 4, -5, 1])

fig,ax = plt.subplots()
ax.plot(y, marker = "o", color = "c", linestyle = ' ')

ax.set_title("Il mio secondo grafico")
ax.set_xlabel("X")
ax.set_ylabel("Y", fontsize = 12)
ax.tick_params(axis = "both", labelsize = 20)
plt.show()
```

```python
x = np.random.uniform(-10, 10, size=50)
y = np.random.uniform(-10, 10, size=50)

colors = np.random.uniform(0, 1, size=(50,3))
diameters = np.random.uniform(50, 150, size = 50)

fig, ax = plt.subplots()
ax.scatter(x, y, c=colors, s=diameters, marker="o")

ax.set_title("Il mio terzo grafico")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid(True)
plt.show()
```

```python
d1 = np.random.randint(1,7,5000)
d2 = np.random.randint(1,7,5000)
somma = d1 + d2

fig,ax = plt.subplots()
ax.hist(somma, bins=11)

ax.set_title("Il mio quarto grafico")
ax.set_xlabel("X")
ax.set_ylabel("Y", fontsize = 12)
ax.tick_params(axis = "both", labelsize = 20)
plt.show()
```
