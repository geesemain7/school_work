*Determine the equation of the spherical surface $S$, with its center point on the line $r: \begin{cases} x = t \\ y = t \\ z = t \end{cases}, t \in \R$ tangent to the plan $\Pi: 3x - y -2z + 14 = 0$ on its point $T = (-4; 0; 1)$.*

<br>

From the plane

$$\Pi: 3x - y - 2z + 14 = 0$$

we can determine its parameters and therefore also its normal vector:

$$\Pi: ax + by + cz + d = 0$$
$$a = 3, b = -1, c = -2$$
$$\overrightarrow{n} (a; b; c) = (3; -1; -2)$$

Given that we need a tangent line, its vector will be identical to that of the plane's normal. Therefore, we can take the parametrical equation of a line to form a tangent line $\gamma$. All we need to do is substitue $l, m, n$ with the normal's parameters.

<br>

As for the line's origin point, we can just use $T(-4; 0; 1)$.

$$
\gamma:
\begin{cases}
    x = x_{o} + l k \\
    y = y_{o} + m k \\
    z = z_{o} + n k
\end{cases}
= 
\begin{cases}
    x = -4 + 3k \\
    y = -k \\
    z = 1 - 2k
\end{cases}
$$

In order to find the center point of $S$, which we will call $O$, we need to intersect $r$ and $\gamma$, then solve the resulting system.

$$
O
\begin{cases}
    t = -4 + 3k \\
    t = -k \\
    t = 1 - 2k
\end{cases}
$$

$$ -4 + 3k = -k \rightarrow k = 1 $$
$$ t = 1 - 2 = -1 $$

Now that we know the value of $t$ and $k$, we can calculate $O$.

$$ O(t; t; t) = (-1; -1; -1) $$

Now that we found the center point, all we need is also $S$'s radius, which is none other than the distance between $T$ and $O$.

$$
\begin{split}
    R & = \overline{TO} \\
      & = \sqrt{ (x_{T} - x_{C})^{2} + (y_{T} - y_{C})^{2} + (z_{T} - z_{C})^{2} } \\
      & = \sqrt{ (-4 + 1)^{2} + (0 + 1)^{2} + (1 + 1)^{2}) } \\
      & = \sqrt{ -3^{2} + 1^{2} + 2^{4} } \\
      & = \sqrt{ 9 + 1 + 4} \\
      & = \sqrt {14}
\end{split}
$$

Now we have everything we need to form the equation for $S$, starting from the generic equation for a spherical surface.

$$
\begin{split}
    S: \quad & (x - x_{o})^{2} + (z - z_{o})^{2} + (z - z_{o})^{2} = R^{2} \\
             & (x - x_{O})^{2} + (z - z_{O})^{2} + (z - z_{O})^{2} = \overline{TO}^{2} \\
            & (x + 1)^{2} + (z + 1)^{2} + (z + 1)^{2} = 14
\end{split}
$$
